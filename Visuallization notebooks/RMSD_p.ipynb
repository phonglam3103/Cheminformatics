```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import seaborn as sns
```

```python
df=pd.read_csv("RMSD_p.csv",delimiter=",",index_col=0)
df
#x=df.set_index('Bước sóng')
```

```python
df.plot(figsize=(13,7),alpha=0.8)
x1,x2,y1,y2 = plt.axis()  
#plt.axis((x1,x2,0,350))
plt.legend(fontsize='large')
#plt.rcParams['figure.dpi'] = 500
plt.xlabel('Time (ns)', fontsize=14)
plt.ylabel('RMSD (nm)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0,100)
plt.legend(ncol=2, loc="lower right")
plt.show()
```

```python

columns= df.columns
plt.rcParams["figure.dpi"] = 300
df[columns[0]].plot(figsize=(5,3),alpha=0.8)
plt.legend(fontsize='large')
x1,x2,y1,y2 = plt.axis()  
plt.axis((0,100,0,0.3))
plt.xlabel('Time (ns)', fontsize=10)
plt.ylabel('RMSD (nm)', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(0,100)
plt.legend(loc="lower right")
plt.show()

for x in range(1,12):
    df[[columns[0],columns[x]]].plot(figsize=(5,3),alpha=0.8)
    plt.legend(fontsize='large')
    x1,x2,y1,y2 = plt.axis()  
    plt.axis((0,100,0,0.3))
    plt.xlabel('Time (ns)', fontsize=10)
    plt.ylabel('RMSD (nm)', fontsize=10)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.xlim(0,100)
    plt.legend(loc="lower right")
    plt.show()
```
